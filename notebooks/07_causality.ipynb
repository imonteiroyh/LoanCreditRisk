# Causal Reasoning to Bridge Risk Models and Business Decisions
This notebook provides a strategic overview of how lightweight causal reasoning can be practically applied to bridge predictive credit risk models with actionable business decisions. While the primary focus of credit models is risk ranking, actual business impact depends on understanding how changes to key levers — such as pricing and credit limit — might shape customer behavior and portfolio outcomes. In this context, causal inference offers a way to estimate and reason about the potential effects of different decisions, even when only observational or mocked data is available. The emphasis here is on deploying principled, transparent methods to enhance business judgment, rather than constructing fully prescriptive or experimental frameworks.

## Motivation

Credit risk models are typically designed to predict the probability of default conditional on observed borrower characteristics and an existing policy environment. While this is essential for ranking risk, it is not sufficient to directly answer business questions such as:

* How much profit would we gain by approving more users?
* What is the optimal trade-off between higher risk and higher volume?
* How should pricing or credit limits be adjusted to maximize long-term value?

A key challenge is that decisions themselves influence outcomes. Interest rates affect borrower affordability and repayment behavior. Credit limits affect utilization, exposure, and sometimes even demand. As a result, observed outcomes in historical data are endogenous to the decisions taken by the incumbent policy.

Causal reasoning provides a framework to reason about these feedback loops. Even when full causal identification is not possible, it helps structure assumptions, isolate mechanisms, and avoid naive interpretations of model outputs as direct business impact.

## Counterfactual Framing

The core shift introduced by causal thinking is moving from prediction to counterfactual questions.

Instead of asking:

> “What is the default risk of this borrower?”

We ask:

> “What would happen *if we changed the decision* applied to this borrower?”

Examples relevant to this setting include:

* Approval counterfactuals

  *What if we relaxed approval cutoffs by one score band?*
  
  How would default rates, volume, and total losses change?

* Pricing counterfactuals

  *What if borrowers currently priced as subgrade E were instead priced like C?*

  How would repayment behavior and portfolio performance respond?

* Limit counterfactuals

  *What if we increased credit limits for low-risk borrowers?*

  Would higher utilization and revenue outweigh the additional exposure?

These are fundamentally causal questions: they ask about outcomes under alternative actions that were not historically taken for the same individuals.

## Lightweight Causal Methods (Practical and Defensible)

Given the absence of randomized approvals or pricing experiments in this dataset, we adopt lightweight causal tools that are transparent and assumption-driven, rather than attempting full structural identification.

### Sensitivity Analysis

Sensitivity analysis evaluates how conclusions change under different, plausible assumptions.

For example:

* Assume a range of elasticities of default with respect to monthly payment.
* Recompute portfolio loss, approval volume, and expected return under each assumption.
* Check whether conclusions (e.g., “model A dominates model B”) are robust across that range.

This approach does not claim point identification, but it answers:

> “Are our conclusions fragile or stable under reasonable economic assumptions?”

### Bounds and Conservative Scenarios

When effects cannot be precisely estimated, we can often bound them.

Examples:

* Upper and lower bounds on default impact when increasing credit limits.
* Worst-case vs best-case scenarios for borrowers moving into newly approved regions.
* Conservative assumptions on take-up decline when interest rates increase.

Bounds are particularly useful for governance:

* If even the pessimistic bound looks acceptable, the decision is safer.
* If conclusions flip depending on small assumption changes, caution is warranted.

### Partial Identification and Directional Insights

In many business decisions, directionality is more important than precision.

For instance:

* Higher interest rates almost certainly increase default risk through payment burden.
* Higher credit limits increase exposure, but also revenue potential and customer satisfaction.
* More competitive pricing likely increases application volume, not just approval rate.

Even without exact magnitudes, partial identification allows us to:

* identify regions of potential upside,
* detect where policy changes are likely dominated (high risk, low growth),
* and focus experimentation where uncertainty and upside are both high.

## Pricing as a Causal Lever

Pricing is one of the clearest examples of a decision variable with causal impact.

### Mechanisms:

* Higher rates → higher monthly payments → increased repayment stress → higher default probability.
* Lower rates → lower burden → improved repayment, but potentially higher take-up.

A causal perspective separates:

* Risk effect: how default probability changes with rate.
* Demand effect: how applicant volume and acceptance change with rate.
* Revenue effect: how interest income trades off with losses.

In practice, we can:

* model default sensitivity to payment changes (as done earlier),
* assume or externally calibrate demand elasticity,
* simulate portfolio-level outcomes under alternative pricing curves.

This enables construction of a risk–return frontier, rather than evaluating models solely on default rate.

## Credit Limits as a Causal Lever

Credit limits introduce a different but equally important causal channel.

### Mechanisms:

* Higher limits → higher utilization and exposure.
* Potential increase in revenue via interest and fees.
* Possible behavioral effects (over-borrowing, risk concentration).

Causal questions include:

* For low-risk borrowers, does increasing limits reduce churn and increase LTV?
* For mid-risk borrowers, where does incremental exposure stop paying off?
* How does limit policy interact with pricing and approval thresholds?

Here again, lightweight causal reasoning helps:

* simulate exposure-weighted loss rather than binary default,
* explore asymmetric policies (e.g., higher limits only for top bands),
* identify segments where limit expansion is dominated by risk.

## From Risk Models to Business Optimization

Combining risk models with causal reasoning allows us to move from:

> “Which model predicts default better?”

to:

> “Which policy, powered by which model, moves the company to a better point on the risk–growth frontier?”

Key outputs of this framework include:

* identification of policy levers with highest upside (pricing vs limits vs approvals),
* understanding where competition effects matter most (volume vs margin),
* prioritization of experiments that are both safe and informative.


## Important Note on Interpretation

> This analysis is exploratory and directional, not prescriptive.

The objective is not to claim precise counterfactual profits, but to:

* structure decisions coherently,
* make assumptions explicit,
* and guide experimentation toward the highest-value and lowest-risk areas.

In environments without full randomization, this approach reflects how causal reasoning is most effectively applied in real credit systems: as a decision-support tool, not a source of false certainty.
